```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_context('talk')
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.countplot(x="species",
              data=penguins,
              palette=["#DA3A27", "lightgrey", "lightgrey"],
              ax=ax)

for i, graph in enumerate(ax.patches) :
    if i > 0 :
        rgba = list(graph.get_facecolor())
        rgba[-1] = 0.5
        graph.set_facecolor(rgba)
    height = graph.get_height()
    ax.text(x=graph.get_x() + 0.31,
            y=height - 10,
            s=height,
            fontweight="bold",
            c="white",
            fontsize=20
           )

ax.spines[["top", "right", "left"]].set_visible(False)
ax.grid(axis="y", color="dimgrey", linewidth=0.25)
ax.yaxis.set_zorder(-1)
ax.set(xlabel="", ylabel="")
ax.tick_params(axis="y", length=0, pad=10)
ax.set_title("Penguins Counts by Species", fontsize=25, fontweight='bold', pad=10, loc='left')
fig.suptitle("* source : seaborn-dataset", x=0.85, y=-0.01, fontsize=15, color='dimgray')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.countplot(y="sex",
              data=penguins,
              palette=["#DA3A27", "lightgrey"],
              ax=ax)

for i, graph in enumerate(ax.patches) :
    if i > 0 :
        rgba = list(graph.get_facecolor())
        rgba[-1] = 0.5
        graph.set_facecolor(rgba)
    width = graph.get_width()
    ax.text(x=width - 15,
            y=graph.get_y() + 0.40,
            s=width,
            fontweight="bold",
            c="white",
            fontsize=20
           )

ax.spines[["top", "right", "bottom"]].set_visible(False)
ax.grid(axis="x", color="dimgrey", linewidth=0.25)
ax.yaxis.set_zorder(-1)
ax.set(xlabel="", ylabel="")
ax.tick_params(axis="x", length=0, pad=10)
ax.set_title("Penguins Counts by Sex", fontsize=25, fontweight='bold', pad=20, loc='left')
fig.suptitle("* source : seaborn-dataset", x=0.80, y=-0.01, fontsize=15, color='dimgray')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.boxplot(x="species",
               y='bill_length_mm',
              data=penguins,
              palette=["lightgrey", "lightgrey", "#DA3A27"],
              ax=ax)

for i, graph in enumerate(ax.patches) :
    if i < 2 :
        rgba = list(graph.get_facecolor())
        rgba[-1] = 0.3
        graph.set_facecolor(rgba)

ax.spines[["top", "right", "left"]].set_visible(False)
ax.grid(axis="y", color="dimgrey", linewidth=0.25)
ax.yaxis.set_zorder(-1)
ax.set(xlabel="", ylabel="")
ax.tick_params(axis="y", length=0, pad=10)
ax.set_title("Penguins bill_length by Species", fontsize=25, fontweight='bold', pad=25, loc='left')
fig.suptitle("* source : seaborn-dataset", x=0.85, y=-0.01, fontsize=15, color='dimgray')
fig.tight_layout()
```
